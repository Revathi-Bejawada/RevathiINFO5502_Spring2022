```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

```python

def getting_month(seriesList):
    
    months_names = 'January February March April May June July August September October November December'.split()
    get_month = seriesList[0].split()[0]
    return months_names.index(get_month)+1

def getting_year(seriesList):
    get_year = int(seriesList[0].split()[1])
    return get_year
    

df['month'] = df[['date']].apply(getting_month, axis=1)
df['year'] = df[['date']].apply(getting_year, axis=1)
```

```python
df.drop('date', axis=1, inplace=True)
df.head()
```

```python
get_range = df[(df['year'] >= 2013) & (df['year'] < 2021)]
#since we want to visualize the data between the years 2013 and 2021 we have taken >= 2013 and < 2021
```

```python
plt.figure(figsize=(15,6))
ax = sns.countplot(get_range.groupby('year').size().sort_values(ascending=False))
plt.xlabel('Num of TED Talks')
plt.title("Number of TED Talks per Month 2013-2020")
#ax.set_yticklabels(['0','100', '200', '300', '400', '500'])



```

```python
# Your python code here:

import yake
import pandas as pd
import matplotlib.pyplot as plt
input = pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv")
input.head()
```

```python
import yake 
num = 1
#to find the top N keywords
for annual in input['year'].unique():
    print(annual)
    size = 1
    n_value = 5 #how many values we want to extract 
    text = " ".join(title for title in input[input['year']==annual]['title'].values)  #formatting
    #in yake we have KeywordExtractor and extract_keywords we are using these to extract the keywords
    to_get_keysfromArticles = yake.KeywordExtractor(top=n_value, features=None)
    extracted_keywords = to_get_keysfromArticles.extract_keywords(text)
  
    for output in extracted_keywords:
        print(output)
    print()
    num += 1
```

```python

import matplotlib.pyplot as plt
input.dropna(axis=0,  inplace=True)
input['year'].astype('int', errors='ignore')

plt.figure(figsize=(15, 15))
for year in input.year.unique():

    text_format = " ".join(title for title in input[input['year']==year]['title'].values)
    to_get_keysfromArticles = yake.KeywordExtractor(n=1, top=1, features=None)
    extracted_keywords = to_get_keysfromArticles.extract_keywords(text_format)
    Scores = [y for x, y in extracted_keywords]
    Keywords =[x for x, y in extracted_keywords]
    plt.scatter(Keywords,Scores,label=year)

plt.xlabel("Keywords", size =10)
plt.ylabel("Scores", size = 10)
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)
plt.rc('font', size=5)
plt.xticks(rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title("Visualization of Keywords", size = 20)
plt.show()
```
